```python
# let's import our clustering library!
import wwdk
```

```python
import numpy as np
data = np.random.randn(100, 3)
cluster = wwdk.cluster.Kmeans(k=3)
result = cluster.fit(data)
```

```python
from matplotlib import pyplot as plt
plt.scatter(data[:, 0], data[:, 1], c=result.labels_, cmap="tab10")
```
