```python
### importieren
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
from sklearn.datasets.samples_generator import make_blobs #random data generator
import pandas as pd
#init set per default to kmeans++ --> define init='random' 
from sklearn.cluster import KMeans
from sklearn.cluster import MiniBatchKMeans 
```

```python
#generate data
X, y_true = make_blobs(n_samples=1500, centers=12,
                       cluster_std=0.5, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

```python
#vanilla
kmeans = KMeans(n_clusters=12, init='random')
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=100, alpha=1);
```

```python
#kmeans++ init
#vanilla
kmeans = KMeans(n_clusters=12)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=100, alpha=1);
```

```python
#minibatches
clustering = MiniBatchKMeans(n_clusters=12, random_state=0, batch_size=100)
clustering.fit(X)
y = clustering.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='viridis')

centers = clustering.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=100, alpha=1);
```
